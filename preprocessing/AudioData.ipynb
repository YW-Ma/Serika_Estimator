```python
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt
import json
import glob
import pickle
import os
import time
import re
```

```python
# create the hashtag set
json_path = '.\json'
json_files = os.listdir(json_path)
hashtag_dict = {}
json_dict = {}
for file in json_files:
    hashtag_dict[file[0:11]] = 1
    json_dict[file[0:17]] = 1
```

```python
# clean wav data
wav_path = '.\wav_files'
wav_files = os.listdir(wav_path)
for file in wav_files:
    if not json_dict.__contains__(file[0:17]):
        command = 'copy ' + wav_path + '\\' + file + ' ' + '.\\deleted\\' + file
        os.system(command)
        os.remove(wav_path + '\\' + file)
        print("removed " + file)
```

```python
# clean json data
wav_path = '.\wav_files'
wav_files = os.listdir(wav_path)
wav_dict = {}
for file in wav_files:
    wav_dict[file[0:17]] = 1

json_path = '.\json'
json_files = os.listdir(json_path)
for file in json_files:
    if not wav_dict.__contains__(file[0:17]):
        command = 'copy ' + json_path + '\\' + file + ' ' + '.\\deleted\\' + file
        os.system(command)
        os.remove(json_path + '\\' + file)
        print("removed " + file)
```

```python
path = r'.\wav'
folders= os.listdir(path)

counter = 0
start_id = 0
for name in folders: # only choose half people
    counter += 1
    if counter <= start_id: # skip processed folders
        continue
    
    dir = path + '\\' +  str(name)
    sub_folders = os.listdir(dir)
    
    for hash_tag in sub_folders: # only choose half vedios
        if not hashtag_dict.__contains__(hash_tag):
            continue
        dir = path + '\\' +  str(name) + '\\' + str(hash_tag)
        sub_sub_folders = os.listdir(dir)
        
        limit = min(len(sub_sub_folders), 3)
        if len(sub_sub_folders) < 10: # pick 1 2 3
            m = 0
            for sentence in sub_sub_folders:  # only choose up to 3 sentence for each vedio
                m += 1
                if m > limit:
                    break
                source_file_path = path + '\\' +  str(name) + '\\' + str(hash_tag) + '\\' + str(sentence)
                sentence_id = hash_tag + '-' + sentence.zfill(5)
                output_file_path = '.\\wav_files\\' + sentence_id
                command = 'copy ' + source_file_path + ' ' + output_file_path
                os.system(command)
                
        elif len(sub_sub_folders) == 10: # pick 1 10 2
            m = 0
            for sentence in sub_sub_folders:  # only choose up to 3 sentence for each vedio
                m += 1
                if m == 1 or m == 2 or m == 10:
                    source_file_path = path + '\\' +  str(name) + '\\' + str(hash_tag) + '\\' + str(sentence)
                    sentence_id = hash_tag + '-' + sentence.zfill(5)
                    output_file_path = '.\\wav_files\\' + sentence_id
                    command = 'copy ' + source_file_path + ' ' + output_file_path
                    os.system(command)
                    
        elif len(sub_sub_folders) > 10: # pick 1 10 11
            m = 0
            for sentence in sub_sub_folders:  # only choose up to 3 sentence for each vedio
                m += 1
                if m == 1 or m == 10 or m == 11:
                    source_file_path = path + '\\' +  str(name) + '\\' + str(hash_tag) + '\\' + str(sentence)
                    sentence_id = hash_tag + '-' + sentence.zfill(5)
                    output_file_path = '.\\wav_files\\' + sentence_id
                    command = 'copy ' + source_file_path + ' ' + output_file_path
                    os.system(command)
```
